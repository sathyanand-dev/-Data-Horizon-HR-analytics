```python
for i in df.columns:
    sns.histplot(df[i])
    plt.show()
```

```python
for i in df.columns:
    if len(list(df[i].unique())) < 7:
        plt.bar(df[i].unique() , df[i].value_counts())
        plt.title(i)
        plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

scaler = LabelEncoder()
df['Performance rating'] = scaler.fit_transform(df['Performance rating'])
df.head()
```

```python
corr = df[['Task given', 'Task done',
       'On time submission', 'Done late', 'Performance rating']].corr()
plt.figure(figsize = (5 , 5))
sns.heatmap(corr , annot = True)
plt.show()
```
